# Online Food Delivery Database Assignment
This notebook generates a dataset, performs normalization, and demonstrates relational algebra concepts.

## Step 1: Generate Sample Dataset

```python
import pandas as pd
df = pd.read_csv('OrderFacts_sample_dataset.csv')
df.head()
```

## Step 2: First Normal Form (1NF)
Ensure atomic attributes and remove repeating groups.

```python
df_1nf = df.copy()
df_1nf.head()
```

## Step 3: Second Normal Form (2NF)
Remove partial dependencies from composite key.

```python
customers = df[['customer_id','customer_name','customer_phone']].drop_duplicates()
restaurants = df[['restaurant_id','restaurant_name']].drop_duplicates()
menu_items = df[['item_id','item_name','item_price','restaurant_id']].drop_duplicates()

orders = df[['order_id','order_date','customer_id','restaurant_id','driver_name']]
order_items = df[['order_id','item_id','quantity']]

customers.head()
```

## Step 4: Third Normal Form (3NF)
Remove transitive dependencies.

```python
delivery = df[['order_id','driver_name']]
delivery.head()
```

## Step 5: Example Relational Algebra Queries

1. Orders by a specific customer  
σ(customer_id = 1)(Orders)

2. Menu items with price > 200  
σ(price > 200)(MenuItem)

3. Join Orders and Customers  
Orders ⨝ Customers

4. Item names with quantities  
π(item_name, quantity)(MenuItem ⨝ OrderItems)

5. Total quantity per order  
γ(order_id, SUM(quantity))(OrderItems)


## Step 6: Pandas Operations vs Relational Algebra

```python
# Merge (Join)
merged = orders.merge(customers,on='customer_id')
merged.head()
```

```python
# Groupby + Aggregate
order_totals = order_items.groupby('order_id')['quantity'].sum()
order_totals.head()
```

```python
# Filter (Selection)
expensive_items = menu_items[menu_items['item_price'] > 200]
expensive_items
```
